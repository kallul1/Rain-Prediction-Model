## Importing Liabraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
df =pd.read_csv("weatherAUS.csv")
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
df=df.drop(["Evaporation","Sunshine","Cloud9am","Cloud3pm","Location","Date"], axis = 1)
df.head()
```

```python
df=df.dropna(axis = 0)
df.shape
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['WindGustDir'] = le.fit_transform(df['WindGustDir'])
df['WindDir9am'] = le.fit_transform(df['WindDir9am'])
df['WindDir3pm'] = le.fit_transform(df['WindDir3pm'])
df['RainToday'] = le.fit_transform(df['RainToday'])
df['RainTomorrow'] = le.fit_transform(df['RainTomorrow'])
```

```python
df.head()
```

```python
X = df.drop(['RainTomorrow'], axis = 1)
Y = df['RainTomorrow']
```

```python
X.head()
```

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x = 'MaxTemp', y='MinTemp', hue = 'RainTomorrow', palette = 'inferno', data=df)
```

```python
plt.figure(figsize = (8,8))
sns.heatmap(df.corr())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

## Linear Regression

```python
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
prediction1 = LR.predict(X_test)
```

```python
mse = mean_squared_error(y_test,prediction1)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,prediction1)
r2 = r2_score(y_test,prediction1)
```

```python
print("mean squared error: ",mse)
print("root mean squared error: ",rmse)
print("mean absolute error: ",mae)
print("r2 score: ",r2)
```

## Logistic Regression

```python
LOR = LogisticRegression(solver='liblinear')
LOR.fit(X_train,y_train)
```

```python
prediction2 = LOR.predict(X_test)
print(confusion_matrix(y_test,prediction2))
print(classification_report(y_test,prediction2))
print(accuracy_score(y_test,prediction2))
```
